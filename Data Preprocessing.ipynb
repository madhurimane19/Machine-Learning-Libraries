# Missing value imputation by mean, median

```python
data.shape
```

```python
data.head()
```

```python
data.tail()
```

```python
missing_val = data.isnull().sum()/data.shape[0] * 100
missing_val
```

```python
missing_val_columns = missing_val[missing_val > 20].keys()
missing_val_columns
```

```python
df2_drop = data.drop(columns = missing_val_columns)
df2_drop.shape
```

```python
df2_num = df2_drop.select_dtypes(include =['int64', 'float64'])
df2_num.head()
```

```python
plt.figure(figsize =(16,9))
sns.heatmap(df2_num.isnull())
```

```python
df2_num.isnull().sum()
```

```python
missing_num_var = [ var for var in df2_num.columns if df2_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()

for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df2_num[var], bins = 20, kde_kws={'linewidth': 5, 'color':'#DC143C'})
warnings.filterwarnings('ignore')
```

```python
#fill all null values with mean value
df2_num_mean= df2_num.fillna(df2_num.mean())
df2_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()

for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df2_num[var], bins = 20, kde_kws={'linewidth': 8, 'color':'red'},label = 'Original')
    sns.distplot(df2_num[var], bins = 20, kde_kws={'linewidth': 5, 'color':'green'}, label = 'Mean')
    plt.legend()
warnings.filterwarnings('ignore')
```

```python
df3_num_median= df2_num.fillna(df2_num.median())
df3_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df2_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df2_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(df3_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
    plt.legend()
warnings.filterwarnings('ignore')
```

```python
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(data[var])
    plt.subplot(3,1,2)
    sns.boxplot(df2_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df3_num_median[var])
warnings.filterwarnings('ignore')    
```

```python
data_concat = pd.concat([df2_num[missing_num_var],df2_num_mean[missing_num_var],df3_num_median[missing_num_var]], axis=1)
```

```python
# original values and mean and medin values in dataset
data_concat[data_concat.isnull().any(axis=1)]
```

# Median

```python
df_copy_median = data.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in data[cat_var].unique():
        df_copy_median.update(data[data.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,data[data.loc[:,cat_var] == var_class][num_var_miss].median()))
```

```python
df_copy_median[num_vars_miss].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(data[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original")
    sns.distplot(df_copy[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean")
    sns.distplot(df_copy_median[var], bins=20, kde_kws={'linewidth':3, 'color':'k'},label="Median")
    plt.legend()
warnings.filterwarnings('ignore')    
    
```

```python
for i, var in enumerate(num_vars_miss):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(data[var])
    plt.subplot(3,1,2)
    sns.boxplot(df_copy[var])
    plt.subplot(3,1,3)
    sns.boxplot(df_copy_median[var])
warnings.filterwarnings('ignore')
```
